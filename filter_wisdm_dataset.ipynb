```python
import pandas as pd
```

```python
wisdm_filename = './all_wisdm.csv'
wisdm_dataset = pd.read_csv(wisdm_filename)

ids = wisdm_dataset['PERSON_ID'].unique()

length = len(ids)
train_ids = ids[0:int(length*0.7)]
validation_ids = ids[int(length*0.7):int(length*0.9)]
test_ids = ids[int(length*0.9):]

wisdm_train_df = wisdm_dataset[wisdm_dataset['PERSON_ID'].isin(train_ids)]
wisdm_validation_df = wisdm_dataset[wisdm_dataset['PERSON_ID'].isin(validation_ids)]
wisdm_test_df = wisdm_dataset[wisdm_dataset['PERSON_ID'].isin(test_ids)]
```

```python
wisdm_train_df = wisdm_train_df[::20]
wisdm_validation_df = wisdm_validation_df[::20]
wisdm_test_df = wisdm_test_df[::20]
```

```python
wisdm_train_df.to_csv('./wisdm_train.csv', mode='w+', index=False)
wisdm_validation_df.to_csv('./wisdm_validation.csv', mode='w+', index=False)
wisdm_train_df.to_csv('./wisdm_test.csv', mode='w+', index=False)
```
